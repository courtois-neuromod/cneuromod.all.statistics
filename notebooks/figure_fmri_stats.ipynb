```python
import os
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.use('Agg')

output_dir = Path(os.environ['OUTPUT_DATA_DIR'])
df = pd.read_csv(output_dir / 'fmri_stats.tsv', sep='\t')
```

```python
datasets = df['dataset'].tolist()

metrics = [
    ('total_runs',             'Total runs',                  'Runs'),
    ('avg_runs_per_session',   'Avg runs / session',          'Runs'),
    ('avg_run_duration_h',     'Avg run duration',            'Hours'),
    ('avg_session_duration_h', 'Avg session duration',        'Hours'),
    ('total_duration_h',       'Total duration',              'Hours'),
]

n_metrics = len(metrics)
n_cols = 3
n_rows = (n_metrics + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 5, n_rows * 4), constrained_layout=True)
axes_flat = axes.flatten()

for i, (col, title, ylabel) in enumerate(metrics):
    ax = axes_flat[i]
    values = df[col].tolist()
    bars = ax.bar(range(len(datasets)), values)
    ax.set_xticks(range(len(datasets)))
    ax.set_xticklabels(datasets, rotation=45, ha='right')
    ax.set_title(title, fontweight='bold')
    ax.set_ylabel(ylabel)

for ax in axes_flat[n_metrics:]:
    ax.set_visible(False)

fig.suptitle('CNeuroMod: fMRI run statistics per dataset', fontsize=14, fontweight='bold')
fig.savefig(output_dir / 'figure_fmri_stats.png', dpi=150)
plt.show()
print(f"Figure saved to {output_dir / 'figure_fmri_stats.png'}")
```
