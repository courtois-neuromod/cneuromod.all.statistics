```python
import os
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.use('Agg')

output_dir = Path(os.environ['OUTPUT_DATA_DIR'])
df = pd.read_csv(output_dir / 'fmri_stats_per_subject.tsv', sep='\t')
```

```python
subjects = sorted(df['subject'].unique())
datasets = sorted(df[df['dataset'] != 'all']['dataset'].unique())
n_datasets = len(datasets)

metrics = [
    ('total_runs',             'Total runs',              'Runs'),
    ('total_duration_h',       'Total duration',          'Hours'),
    ('avg_session_duration_h', 'Avg session duration',    'Hours'),
    ('total_volumes',          'Total volumes',           'Volumes'),
    ('avg_volumes_per_session','Avg volumes per session', 'Volumes'),
]

for metric, metric_label, ylabel in metrics:
    n_cols = 4
    n_panels = 1 + n_datasets
    n_rows = (n_panels + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 3.5, n_rows * 3), constrained_layout=True)
    axes_flat = axes.flatten()

    # Summary panel: use the "all" dataset rows (cumulative per subject)
    ax = axes_flat[0]
    all_df = df[df['dataset'] == 'all'].set_index('subject')
    ax.bar(subjects, [all_df.loc[s, metric] if s in all_df.index else 0 for s in subjects])
    ax.set_title('Total (all datasets)', fontweight='bold')
    ax.set_xlabel('Subject')
    ax.set_ylabel(ylabel)
    ax.tick_params(axis='x', rotation=45)

    # One panel per dataset (exclude "all")
    for i, dataset in enumerate(datasets):
        ax = axes_flat[i + 1]
        sub_df = df[df['dataset'] == dataset].set_index('subject')
        values = [sub_df.loc[s, metric] if s in sub_df.index else 0 for s in subjects]
        ax.bar(subjects, values)
        ax.set_title(dataset)
        ax.set_xlabel('Subject')
        ax.set_ylabel(ylabel)
        ax.tick_params(axis='x', rotation=45)

    for ax in axes_flat[n_panels:]:
        ax.set_visible(False)

    fig.suptitle(f'CNeuroMod: {metric_label} per subject per dataset', fontsize=14, fontweight='bold')
    out_path = output_dir / f'figure_fmri_per_subject_{metric}.png'
    fig.savefig(out_path, dpi=150)
    plt.show()
    print(f"Figure saved to {out_path}")
```
