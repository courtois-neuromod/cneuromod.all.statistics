```python
import os
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.use('Agg')

output_dir = Path(os.environ['OUTPUT_DATA_DIR'])
df = pd.read_csv(output_dir / 'session_counts.tsv', sep='\t')
```

```python
subjects = sorted(df['subject'].unique())
datasets = sorted(df['dataset'].unique())
n_datasets = len(datasets)

# Layout: 1 summary panel + one panel per dataset, up to 4 columns
n_cols = 4
n_panels = 1 + n_datasets
n_rows = (n_panels + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 3.5, n_rows * 3), constrained_layout=True)
axes_flat = axes.flatten()

# Summary panel: total sessions per subject across all datasets
ax = axes_flat[0]
totals = df.groupby('subject')['n_sessions'].sum()
ax.bar(subjects, [totals[s] for s in subjects])
ax.set_title('Total (all datasets)', fontweight='bold')
ax.set_xlabel('Subject')
ax.set_ylabel('Sessions')
ax.tick_params(axis='x', rotation=45)

# One panel per dataset
for i, dataset in enumerate(datasets):
    ax = axes_flat[i + 1]
    sub_df = df[df['dataset'] == dataset].set_index('subject')
    counts = [sub_df.loc[s, 'n_sessions'] if s in sub_df.index else 0 for s in subjects]
    ax.bar(subjects, counts)
    ax.set_title(dataset)
    ax.set_xlabel('Subject')
    ax.set_ylabel('Sessions')
    ax.tick_params(axis='x', rotation=45)

# Hide unused panels
for ax in axes_flat[n_panels:]:
    ax.set_visible(False)

fig.suptitle('CNeuroMod: sessions per subject per dataset', fontsize=14, fontweight='bold')
fig.savefig(output_dir / 'figure_sessions.png', dpi=150)
plt.show()
print(f"Figure saved to {output_dir / 'figure_sessions.png'}")
```
